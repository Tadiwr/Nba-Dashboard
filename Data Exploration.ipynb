```python
from data.repo import Repo as rp
import pandas as pd

teams = []
streaks = []

for x in range(1, 31):
    teamData = rp.teams_api.getTeamData(x)
    teams.append(teamData.name)
    streaks.append(teamData.stats.win_percentage)

df = pd.DataFrame({
    "Team":teams,
    "Streak":streaks
})

df
```
